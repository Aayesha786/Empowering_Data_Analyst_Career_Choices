# 6) Methods and Tools Used for Removing Outliers with Boxplot:

### Pandas: Pandas is used for handling the DataFrame and data manipulation.

### NumPy: NumPy is utilized for calculations involving arrays and mathematical operations.

### SciPy (stats module): Specifically, the stats.zscore function from SciPy is applied to calculate Z-scores.

### Boxplot: A boxplot is used to visualize the distribution of data and identify potential outliers.

### Z-scores: Z-scores are statistical measures that determine how far a data point deviates from the mean of a dataset.

### Threshold: A threshold value (in this case, 3) is set to determine when a data point is considered an outlier.

### Boolean Mask: A boolean mask is created based on the Z-scores and threshold to identify rows with outliers.

### Data Filtering: The boolean mask is applied to the original DataFrame to filter out rows containing outliers.

### Boxplot Visualization: A boxplot is used to visually inspect the distribution of data, identify outliers, and confirm the effectiveness of the outlier removal process.

# Checking the Outliers in the Salary column using the Boxplot

```python
import matplotlib.pyplot as plt
plt.boxplot(df["Salary in USD"])
plt.title("Box plot salaries(whis=1.5)")
plt.show()
```

### We probaly got an important number of outliers in salaries.

```python
import matplotlib.pyplot as plt

# Create a boxplot without outliers
plt.boxplot(df["Salary in USD"], showfliers=False)
plt.title("Box plot salaries (Outliers Removed)")
plt.show()
```

```python
# import pandas as pd
import numpy as np
from scipy import stats

# Calculate the Z-scores for each numeric column in your DataFrame
z_scores = np.abs(stats.zscore(df.select_dtypes(include=np.number)))

# Define a threshold (e.g., 3) beyond which data points will be considered outliers
threshold = 3

# Create a boolean mask to identify rows with outliers based on the Z-scores
outlier_mask = (z_scores > threshold).any(axis=1)

# Use the mask to filter out rows with outliers from your DataFrame
df = df[~outlier_mask]

# Now, the original DataFrame (df) contains your data with outliers removed based on the Z-score method
df
```

# 8) "Which data analyst roles are paid the highest in the US?"

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the median salaries for different job titles in the United States
median_salary_US_ds = df[(df['Job Title'] == 'Data Scientist') & (df['Employee Residence'] == 'United States')]['Salary in USD'].median()
median_salary_US_mle = df[(df['Job Title'] == 'Machine Learning Engineer') & (df['Employee Residence'] == 'United States')]['Salary in USD'].median()
median_salary_US_da = df[(df['Job Title'] == 'Data Analyst') & (df['Employee Residence'] == 'United States')]['Salary in USD'].median()
median_salary_US_de = df[(df['Job Title'] == 'Data Engineer') & (df['Employee Residence'] == 'United States')]['Salary in USD'].median()

# Create a DataFrame for easier sorting
data = {
    'Job Title': ['Data Scientist', 'Machine Learning Engineer', 'Data Analyst', 'Data Engineer'],
    'Median Salary in USD': [median_salary_US_ds, median_salary_US_mle, median_salary_US_da, median_salary_US_de]
}
median_salary_df = pd.DataFrame(data)

# Sort the DataFrame by median salary in descending order
median_salary_df = median_salary_df.sort_values(by='Median Salary in USD', ascending=False)

# Create a bar chart to visualize the sorted median salaries for different job titles in the United States
job_titles = median_salary_df['Job Title']
median_salaries = median_salary_df['Median Salary in USD']

plt.figure(figsize=(10, 6))
plt.bar(job_titles, median_salaries, color='skyblue')
plt.xlabel('Job Title')
plt.ylabel('Median Salary in USD')
plt.title('Median Salaries for Different Job Titles in the United States (Sorted by Salary)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

```

### The data presents a noteworthy finding: the roles of "Machine Learning Engineer" and "Data scientist" command the highest median salaries in the United States, with figures hovering around $175,000 to 150000.

### Machine Learning Engineers, with their expertise in building advanced predictive models and machine learning algorithms, are evidently highly sought after. Their role in creating and deploying cutting-edge AI solutions positions them as crucial contributors in data-driven industries.
